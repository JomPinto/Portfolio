```python
def salary_seperatorinator(salary):
    ##str = "₱40,000 - ₱80,000"
    str2 = (salary.replace('₱', ''))
    str3 = (str2.replace(',', ''))
    stroutput = [int(s) for s in str3.split() if s.isdigit()]
    if not len(stroutput):
        return salary
    else:
        return stroutput[0]
```
