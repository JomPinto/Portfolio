```python
def date_creatorinator(posted_ago):
    posted_ago2 = (posted_ago.replace('Posted On: ', ''))
    str = posted_ago2
    date_checker = str.find("day")
    if date_checker > 0:
        num_posted = [int(s) for s in str.split() if s.isdigit()]
        day = int(today.strftime("%d")) - int(num_posted[0])
        month = today.strftime("%m")
        year =  today.strftime("%Y")
        #print (month + " -"  , day , "- " + year)
        return ("{}-{}-{}").format(year,day,month)
    else:
        if str.find("a") == 0:
            str = (str.replace('a', '1'))
        num_posted = [int(s) for s in str.split() if s.isdigit()]
        month = int(today.strftime("%m")) - int(num_posted[0])
        day = today.strftime("%d")
        year =  today.strftime("%Y")
        if month < 1:
            return ("{}-{}-{}").format(year,day,month+12)
            #print (year , "-"  , day , "-" , month+12)
        else:
            return ("{}-{}-{}").format(year,day,month)
            #print (year , "-"  , day , "-" , month)
```

```python
def salary_seperatorinator_monster(salary):
    str2 = (salary.replace('₱', ''))
    str3 = (str2.replace(',', ''))
    stroutput = [int(s) for s in str3.split() if s.isdigit()]
    if not len(stroutput):
        return salary
    else:
        return stroutput[0]
```
